```python
# exemplar usage from the documentation
from sentence_transformers import SentenceTransformer, util
sentences = ["I had a nice day", "happy" , "sad"]

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

#Compute embedding for both lists
embedding_1= model.encode(sentences[0], convert_to_tensor=True)
embedding_2 = model.encode(sentences[1], convert_to_tensor=True)
embedding_3 = model.encode(sentences[2], convert_to_tensor=True)

prob_a = util.pytorch_cos_sim(embedding_1, embedding_2)
prob_b =  util.pytorch_cos_sim(embedding_1, embedding_3)

print(prob_a,prob_b)
```

```python
prob_a > prob_b
```

```python
import numpy as np , pandas as pd
import seaborn as sns
```

```python
cupid = pd.read_csv("../okcupid.csv")
```

```python
def token_bio(sen):
    #Compute embedding for both lists
    try:
        embedding_1= model.encode(sen, convert_to_tensor=True)
    except TypeError:
        return "other"
    
    embedding_2 = model.encode("academic", convert_to_tensor=True)
    embedding_3 = model.encode("adventurous", convert_to_tensor=True)

    prob_a = util.pytorch_cos_sim(embedding_1, embedding_2)
    prob_b =  util.pytorch_cos_sim(embedding_1, embedding_3)
    if prob_a > prob_b:
        print("ac")
        return "academic"
    else:
        print("ad")
        return "adventurous"
    
```

```python
from tqdm import tqdm
cupid["pref"] = cupid["essay0"].apply(token_bio)
```
